```python
from tqdm import tnrange
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
import torchvision.transforms as transforms
from torch.autograd import Variable
import matplotlib.pyplot as plt
import numpy as np
import gc
import os
%matplotlib inline  

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
BATCH_SIZE = 1000

# list all transformations
transform = transforms.Compose(
    [transforms.ToTensor(),
     transforms.Normalize((0.1307,), (0.3081,))])

# download and load training dataset
trainset = torchvision.datasets.MNIST(root='./data', train=True,
                                        download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=BATCH_SIZE,
                                          shuffle=True, num_workers=4)

# download and load testing dataset
testset = torchvision.datasets.MNIST(root='./data', train=False,
                                       download=True, transform=transform)
testloader = torch.utils.data.DataLoader(testset, batch_size=BATCH_SIZE,
                                         shuffle=False, num_workers=4)
```

```python
def onehotTensor(category,n_categories):
    tensor = torch.zeros(1, n_categories,dtype=torch.long)
    tensor[0][category] = 1
    return tensor
        
def get_accuracy(logit, target):
    batch_size = len(target)
    corrects = (torch.max(logit, 1)[1].view(target.size()).data == target.data).sum()
    accuracy = 100.0 * corrects/batch_size
    return accuracy.item()

def nparam(ninputs,nhidden,noutputs):
    return ninputs*(nhidden+1) + nhidden*(nhidden+1)+nhidden*(noutputs+1)

# a custom RNN
class PRNN(nn.Module):
    def __init__(self, n_inputs, n_hidden,n_output,dt,device):
        super(PRNN, self).__init__()
        self.n_inputs = n_inputs
        self.n_hidden = n_hidden
        self.n_output = n_output
        self.dt = nn.Parameter(torch.Tensor([dt]))
        self.a = nn.Parameter(torch.Tensor([1]))
        self.sig = nn.Sigmoid()
        self.decoder = nn.Linear(n_hidden, n_output)
        self.encoder = nn.Linear(n_inputs, n_hidden)
        self.recurrent = nn.Linear(n_hidden,n_hidden)
        
    def forward(self, x0):
        x0=x0.permute(1,0,2)
        self.h1 = torch.zeros(1,BATCH_SIZE,self.n_hidden).to(device)
        for i in range(x0.size(0)):
            self.h1 = (1-self.sig(self.dt))*self.h1+self.sig(self.dt)*self.a*torch.tanh(self.encoder(x0[i,:,:])+self.recurrent(self.h1))
        self.y1 = self.decoder(self.h1)
        
        return self.y1
```

```python
# parameters 
N_STEPS = 28
N_INPUTS = 28
N_HIDDEN = 112
N_OUTPUTS = 10
N_EPHOCS = 11
N_REPS = 15
N_PARAMS = nparam(N_INPUTS,N_HIDDEN,N_OUTPUTS)

fracs = np.arange(0.125,1,0.125,dtype=np.float)
N_FRAC = len(fracs)
reg_lambda = 4
```

```python
# regularizing digonal blocks of the partitioned RNN
train_loss_P = np.zeros((N_EPHOCS,N_FRAC,N_REPS))
train_acc_P = np.zeros((N_EPHOCS,N_FRAC,N_REPS))
test_loss_P = np.zeros((N_EPHOCS,N_FRAC,N_REPS))
test_acc_P = np.zeros((N_EPHOCS,N_FRAC,N_REPS))
Phist_P = np.zeros((N_PARAMS,N_EPHOCS,N_FRAC,N_REPS))

model_P = [None]*N_FRAC*N_REPS
regval_P = []
for r in tnrange(N_REPS):
    for k in tnrange(N_FRAC):
        gidx = int(N_HIDDEN*fracs[k])
        model_path = 'D:\chongguo\git\AM221pset\Project\Final Project\History\Frac\model_P_rep_{}_frac_{:d}_1000.pt'.format(r,int(gidx*1000))
        model_P[k+r*N_FRAC] = PRNN(N_INPUTS,N_HIDDEN,N_OUTPUTS,1,device).to(device)
        l2_reg = torch.Tensor(1).to(device)
        optimizer = torch.optim.SGD(model_P[k+r*N_FRAC].parameters(), lr=1e-2, momentum=0.9)
        criterion = nn.CrossEntropyLoss()
        
        for epoch in range(N_EPHOCS):
            nps = 0
            running_train_loss=0
            running_train_acc=0
            model_P[k+r*N_FRAC].train()
            for p, param in enumerate(model_P[k+r*N_FRAC].parameters()):
                if param.requires_grad:
                    plist = torch.flatten(param.data)
                    for j in range(plist.size(0)):
                        Phist_P[nps,epoch,k,r]=plist[j].item()
                        nps+=1

            for i, (x, y_tar) in enumerate(trainloader):
                l2_reg = 0
                x, y_tar = x.to(device), y_tar.to(device)
                optimizer.zero_grad()
                y_pred = model_P[k+r*N_FRAC](x.view(BATCH_SIZE,N_STEPS,N_INPUTS))
                loss = criterion(y_pred.view(BATCH_SIZE,N_OUTPUTS),y_tar)
                for p,param in enumerate(model_P[k+r*N_FRAC].parameters()):
                    if param.requires_grad and len(param.shape)==2:
                        if param.shape[0]==N_HIDDEN and param.shape[1]==N_HIDDEN:
                            l2_reg = l2_reg + param[:gidx,:gidx].norm(p=1)
                            l2_reg = l2_reg + param[gidx:,gidx:].norm(p=1)
                        elif param.shape[1]==N_HIDDEN:
                            l2_reg = l2_reg + param[:,gidx:].norm(p=1)
                        elif param.shape[0]==N_HIDDEN:
                            l2_reg = l2_reg + param[:gidx,:].norm(p=1)
                regval_P.append(l2_reg.item())
                loss = loss + l2_reg*reg_lambda/BATCH_SIZE
                loss.backward()
                optimizer.step()
                running_train_loss+=loss.item()
                running_train_acc+=get_accuracy(y_pred.view(BATCH_SIZE,N_OUTPUTS), y_tar)
            
            running_test_acc=0
            running_test_loss=0
            model_P[k+r*N_FRAC].eval()
            for i,(x_test, y_test_tar) in enumerate(testloader):
                x_test, y_test_tar = x_test.to(device), y_test_tar.to(device)
                y_test_pred = model_P[k+r*N_FRAC](x_test.view(BATCH_SIZE,N_STEPS,N_INPUTS))
                loss = criterion(y_test_pred.view(BATCH_SIZE,N_OUTPUTS),y_test_tar)
                running_test_loss+=loss.item()
                running_test_acc+=get_accuracy(y_test_pred.view(BATCH_SIZE,N_OUTPUTS), y_test_tar)
                
            train_loss_P[epoch,k,r] = running_train_loss/len(trainloader)
            train_acc_P[epoch,k,r] = running_train_acc/len(trainloader)
            test_loss_P[epoch,k,r] = running_test_loss/len(testloader)
            test_acc_P[epoch,k,r] = running_test_acc/len(testloader)
            
        # save the model and free the memory  
        #torch.save(model_P[k+r*N_FRAC].state_dict(), model_path)
        model_P[k+r*N_FRAC] = [None]
        del(l2_reg,loss,optimizer,criterion,plist,param)
```

```python
# regularizing random elements of the matrix
train_loss_R = np.zeros((N_EPHOCS,N_FRAC,N_REPS))
train_acc_R = np.zeros((N_EPHOCS,N_FRAC,N_REPS))
test_loss_R = np.zeros((N_EPHOCS,N_FRAC,N_REPS))
test_acc_R = np.zeros((N_EPHOCS,N_FRAC,N_REPS))
Phist_R = np.zeros((N_PARAMS,N_EPHOCS,N_FRAC,N_REPS))

model_R =  [None] * N_FRAC*N_REPS
regval_R = []
for r in tnrange(N_REPS):
    for k in tnrange(N_FRAC):
        gidx = int(N_HIDDEN*fracs[k])
        model_path = 'D:\chongguo\git\AM221pset\Project\Final Project\History\Frac\model_R_rep_{}_frac_{:d}_1000.pt'.format(r,int(gidx*1000))
        model_R[k+r*N_FRAC] = PRNN(N_INPUTS,N_HIDDEN,N_OUTPUTS,1,device).to(device)
        l2_reg = torch.Tensor(1).to(device)
        optimizer = torch.optim.SGD(model_R[k+r*N_FRAC].parameters(), lr=1e-2, momentum=0.9)
        criterion = nn.CrossEntropyLoss()
        
        # generate a random binary mask for the parameter
        rndmask = torch.FloatTensor(N_HIDDEN, N_HIDDEN).uniform_() < fracs[k]
        while not rndmask.float().norm(p=1)==int(N_HIDDEN**2*fracs[k]):
            if rndmask.float().norm(p=1)>int(N_HIDDEN**2*fracs[k]):
                while True:
                    rndidx = np.random.randint(N_HIDDEN,size=(2))
                    if rndmask[rndidx[0],rndidx[1]]==1:
                        rndmask[rndidx[0],rndidx[1]]=0
                        break
            else:
                while True:
                    rndidx = np.random.randint(N_HIDDEN,size=(2))
                    if rndmask[rndidx[0],rndidx[1]]==0:
                        rndmask[rndidx[0],rndidx[1]]=1
                        break
        rndmask = rndmask.to(device)
        
        for epoch in range(N_EPHOCS):
            nps = 0
            running_train_loss=0
            running_train_acc=0
            model_R[k+r*N_FRAC].train()
            for p, param in enumerate(model_R[k+r*N_FRAC].parameters()):
                if param.requires_grad:
                    plist = torch.flatten(param.data)
                    for j in range(plist.size(0)):
                        Phist_R[nps,epoch,k,r]=plist[j].item()
                        nps+=1
                        
            for i, (x,y_tar) in enumerate(trainloader):
                l2_reg = 0
                x, y_tar = x.to(device), y_tar.to(device)
                optimizer.zero_grad()
                y_pred = model_R[k+r*N_FRAC](x.view(BATCH_SIZE,N_STEPS,N_INPUTS))
                loss = criterion(y_pred.view(BATCH_SIZE,N_OUTPUTS),y_tar)
                for p,param in enumerate(model_R[k+r*N_FRAC].parameters()):
                    if param.requires_grad and len(param.shape)==2:
                        if param.shape[0]==N_HIDDEN and param.shape[1]==N_HIDDEN:
                            l2_reg = l2_reg + torch.masked_select(param,rndmask).norm(p=1)
                        elif param.shape[1]==N_HIDDEN:
                            l2_reg = l2_reg + param[:,gidx:].norm(p=1)
                        elif param.shape[0]==N_HIDDEN:
                            l2_reg = l2_reg + param[:gidx,:].norm(p=1)
                regval_R.append(l2_reg.item())
                loss = loss + l2_reg*reg_lambda/BATCH_SIZE
                loss.backward()
                optimizer.step()
                running_train_loss+=loss.item()
                running_train_acc+=get_accuracy(y_pred.view(BATCH_SIZE,N_OUTPUTS), y_tar)
              
            running_test_acc=0
            running_test_loss=0
            model_R[k+r*N_FRAC].eval()
            for i,(x_test, y_test_tar) in enumerate(testloader):
                x_test, y_test_tar = x_test.to(device), y_test_tar.to(device)
                y_test_pred = model_R[k+r*N_FRAC](x_test.view(BATCH_SIZE,N_STEPS,N_INPUTS))
                running_test_loss+=loss.item()
                running_test_acc+=get_accuracy(y_test_pred.view(BATCH_SIZE,N_OUTPUTS), y_test_tar)
              
            train_loss_R[epoch,k,r] = running_train_loss/len(trainloader)
            train_acc_R[epoch,k,r] = running_train_acc/len(trainloader)
            test_loss_R[epoch,k,r] = running_test_loss/len(testloader)
            test_acc_R[epoch,k,r] = running_test_acc/len(testloader)
            
        # save the model and free the memory  
        #torch.save(model_R[k+r*N_FRAC].state_dict(), model_path)
        model_R[k+r*N_FRAC] = [None]
        del(l2_reg,loss,optimizer,criterion,plist,param)
```

```python
# regularizing off-digonal blocks of the partitioned RNN
train_loss_C = np.zeros((N_EPHOCS,N_FRAC,N_REPS))
train_acc_C = np.zeros((N_EPHOCS,N_FRAC,N_REPS))
test_loss_C = np.zeros((N_EPHOCS,N_FRAC,N_REPS))
test_acc_C = np.zeros((N_EPHOCS,N_FRAC,N_REPS))
Phist_C = np.zeros((N_PARAMS,N_EPHOCS,N_FRAC,N_REPS))

model_C =  [None] * N_FRAC*N_REPS
regval_C = []
for r in tnrange(N_REPS):
    for k in tnrange(N_FRAC):
        gidx = int(N_HIDDEN*fracs[k])
        model_path = 'D:\chongguo\git\AM221pset\Project\Final Project\History\Frac\model_C_rep_{}_frac_{:d}_1000.pt'.format(r,int(gidx*1000))
        model_C[k+r*N_FRAC] = PRNN(N_INPUTS,N_HIDDEN,N_OUTPUTS,1,device).to(device)
        l2_reg = torch.Tensor(1).to(device)
        optimizer = torch.optim.SGD(model_C[k+r*N_FRAC].parameters(), lr=1e-2, momentum=0.9)
        criterion = nn.CrossEntropyLoss()
        
        for epoch in range(N_EPHOCS):
            nps = 0
            running_train_loss=0
            running_train_acc=0
            model_C[k+r*N_FRAC].train()
            for p, param in enumerate(model_C[k+r*N_FRAC].parameters()):
                if param.requires_grad:
                    plist = torch.flatten(param.data)
                    for j in range(plist.size(0)):
                        Phist_C[nps,epoch,k,r]=plist[j].item()
                        nps+=1

            for i, (x, y_tar) in enumerate(trainloader):
                l2_reg = 0
                x, y_tar = x.to(device), y_tar.to(device)
                optimizer.zero_grad()
                y_pred = model_C[k+r*N_FRAC](x.view(BATCH_SIZE,N_STEPS,N_INPUTS))
                loss = criterion(y_pred.view(BATCH_SIZE,N_OUTPUTS),y_tar)
                for p,param in enumerate(model_C[k+r*N_FRAC].parameters()):
                    if param.requires_grad and len(param.shape)==2:
                        if param.shape[0]==N_HIDDEN and param.shape[1]==N_HIDDEN:
                            l2_reg = l2_reg + param[:gidx,gidx:].norm(p=1)
                            l2_reg = l2_reg + param[gidx:,:gidx].norm(p=1)
                        elif param.shape[1]==N_HIDDEN:
                            l2_reg = l2_reg + param[:,gidx:].norm(p=1)
                        elif param.shape[0]==N_HIDDEN:
                            l2_reg = l2_reg + param[:gidx,:].norm(p=1)
                regval_C.append(l2_reg.item())
                loss = loss + l2_reg*reg_lambda/BATCH_SIZE
                loss.backward()
                optimizer.step()
                running_train_loss+=loss.item()
                running_train_acc+=get_accuracy(y_pred.view(BATCH_SIZE,N_OUTPUTS), y_tar)
              
            running_test_acc=0
            running_test_loss=0
            model_C[k+r*N_FRAC].eval()
            for i,(x_test, y_test_tar) in enumerate(testloader):
                x_test, y_test_tar = x_test.to(device), y_test_tar.to(device)
                y_test_pred = model_C[k+r*N_FRAC](x_test.view(BATCH_SIZE,N_STEPS,N_INPUTS))
                running_test_loss+=loss.item()
                running_test_acc+=get_accuracy(y_test_pred.view(BATCH_SIZE,N_OUTPUTS), y_test_tar)
              
            train_loss_C[epoch,k,r] = running_train_loss/len(trainloader)
            train_acc_C[epoch,k,r] = running_train_acc/len(trainloader)
            test_loss_C[epoch,k,r] = running_test_loss/len(testloader)
            test_acc_C[epoch,k,r] = running_test_acc/len(testloader)
            
        # save the model and free the memory  
        #torch.save(model_C[k+r*N_FRAC].state_dict(), model_path)
        model_C[k+r*N_FRAC] = [None]
        del(l2_reg,loss,optimizer,criterion,plist,param)
```
